```python
import time
import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

service_obj = Service("/Users/kavin/Desktop/Sem-2/Retail Analytics/data scraping /web scraping/chromedriver-mac-x64/chromedriver")
driver = webdriver.Chrome(service=service_obj)
driver.get("https://www.flightradar24.com/data/airports/lgw/departures")
time.sleep(6)
driver.find_element("xpath", "//*[@id='onetrust-reject-all-handler']").click()
driver.maximize_window()

max_attempts = 8
attempts = 0

while attempts < max_attempts:
    try:
        button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//*[@class='btn btn-table-action btn-flights-load']")))

        if button.is_enabled():
            button.click()
            time.sleep(3)
        else:
            break
    except:
        attempts += 1

WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, "//*[@class='table table-condensed table-hover data-table m-n-t-15']")))

table = driver.find_element("xpath","//*[@class='table table-condensed table-hover data-table m-n-t-15']")
rows = table.find_elements("xpath",".//tr[not(@class='ng-scope')]")


data = []

for row in rows:
    cols = row.find_elements(By.CSS_SELECTOR, "td")
    if len(cols) >= 6 and "easyJet" in cols[3].text:
        time = cols[0].text
        flight = cols[1].text
        to = cols[2].text
        airline = cols[3].text
        aircraft = cols[4].text
        status = cols[6].text  
        data.append([time, flight, to, airline, aircraft, status])

driver.quit()

departure = pd.DataFrame(data, columns=["TIME", "FLIGHT", "TO", "AIRLINE", "AIRCRAFT", "STATUS"])
departure.to_excel("flight_DepartureData.xlsx", index=False)
departure.head()
```
